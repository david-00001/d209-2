```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import confusion_matrix, roc_curve, auc


from sklearn.tree import DecisionTreeRegressor
```

```python
# Import CSV
full_df = pd.read_csv('churn_clean.csv')
```

```python
# List columns
full_df.columns
```

```python
# Reduce dataset to necessary columns related to research question
columns = ['Area', 'Children', 'Age', 'Income', 'Marital', 'Gender', 'Churn', 'Outage_sec_perweek',
                   'Email', 'Contacts', 'Yearly_equip_failure', 'Techie', 'Contract', 'Port_modem', 
                   'Tablet', 'InternetService', 'Phone', 'Multiple', 'OnlineSecurity', 'OnlineBackup', 
                   'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling',
                   'PaymentMethod', 'Tenure', 'MonthlyCharge', 'Bandwidth_GB_Year', 'Item1', 'Item2', 
                   'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8']

df = full_df[columns]
```

# Check for Null values

```python
# Initial check for null values
df.info()
```

```python
# Check for abnormal representations of null values. Also helps with checking for partial duplicates.
for col in columns:
    print(col)
    print(df[col].value_counts(dropna=False))
    print()
    print()
```

# Check for Complete Duplicates

```python
# Check for complete duplicates
if df.duplicated().any():
    print('Has complete duplicates')
else:
    print('No complete duplicates')
```

# Check for Outliers

```python
# Create boxplot and if outliers exist, calculate using IQR method
for col in columns:
    if df[col].dtype != 'object':
        # Create boxplot
        df.boxplot(column=col)
        plt.title(f'Boxplot of {col}')
        plt.show()
        
        # Calculate IQR lower and upper bounds to determine outliers
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count and display outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        num_outliers = outliers.shape[0]
        print(f"{col} has {num_outliers} outliers")     
        
        
    else:
        # For categorical variables, determine if any category makes up a small percentage
        print()
        print(f'Value counts for {col}')
        print(df[col].value_counts())
        print()
```

# Encoding

```python
# Make a copy because otherwise it produces a pandas slicing warning when we encode
df = df.copy()
```

```python
# Encode yes/no columns with simple replacement
yes_no_columns = ['Churn', 'Techie', 'Port_modem', 'Tablet', 'Phone', 'Multiple', 'OnlineSecurity',
                  'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
                  'PaperlessBilling']

df[yes_no_columns] = df[yes_no_columns].replace({'No': 0, 'Yes': 1})
```

```python
# One-hot encoding for remaining categorical variables (NumFOCUS, 2024)
df_encoded = pd.get_dummies(df, columns=['Area', 'Marital', 'Gender', 'Contract', 'InternetService',
                                         'PaymentMethod'], drop_first=True)
df_encoded.columns
```

# Assign X,Y, Split Data, and StandardScaler

```python
# Assign features to X and label to y (Boorman, n.d.c)
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']
```

```python
# Export prepared data to CSV
df_encoded.to_csv('Prepared Churn.csv', index=False)
```

```python
# Split data using train_test_split (Boorman, n.d.a)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
# Use StandardScaler to scale data because we are using a distance-based algorithm (Filho, 2023)

# Specify columns to scale
columns_to_scale = ['Children', 'Age', 'Income', 'Outage_sec_perweek', 'Email', 'Contacts', 
                    'Yearly_equip_failure', 'Tenure', 'MonthlyCharge', 'Bandwidth_GB_Year', 'Item1',
                    'Item2', 'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8']

#Initialize StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to specified columns in training and test data
X_train_scaled = scaler.fit_transform(X_train[columns_to_scale])
X_test_scaled = scaler.transform(X_test[columns_to_scale])

# Replace original columns with scaled data
X_train[columns_to_scale] = X_train_scaled
X_test[columns_to_scale] = X_test_scaled
```

```python
# Export training and test datasets
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```

# Modeling

```python
# Find optimal value for n_neighbors hyperparameter by first creating an array of possible neighbor-values
# and then create the model using that value as the n_neighbors (Boorman, n.d.b)

# Create empty dictionaries to store accuracy scores
train_accuracies = {}
test_accuracies = {}

# Create an array of neighbor-values
neighbors = np.arange(1,26)

# For each neighbor-value in array
for neighbor in neighbors:
    
    # Initialize and fit the model
    knn = KNN(n_neighbors=neighbor)
    knn.fit(X_train_scaled, y_train)
    
    # Store the accuracy scores
    train_accuracies[neighbor] = knn.score(X_train_scaled, y_train)
    test_accuracies[neighbor] = knn.score(X_test_scaled, y_test)
```

```python
# Plot the accuracies based on the neighbor-values (Boorman, n.d.b)
plt.figure(figsize=(8,6))
plt.title('Accuracy based on Number of Neighbors')
plt.plot(neighbors, train_accuracies.values(), label='Training Accuracy')
plt.plot(neighbors, test_accuracies.values(), label='Testing Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
# After selecting 19 as the optimal neighbor-value, re-initialize and re-fit the model
knn = KNN(n_neighbors=19)
knn.fit(X_train_scaled, y_train)

# Make predictions (Boorman, n.d.b)
y_pred = knn.predict(X_test_scaled)
```

```python
# Use a confusion matrix to assess model accuracy, precision, and recall (Jones, n.d.)

# Create the confusion matrix using the test and prediction labels
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)

# Assign each matrix element to its appropriate name and print the results
TN = conf_matrix[0,0]
FN = conf_matrix[1,0]
FP = conf_matrix[0,1]
TP = conf_matrix[1,0]
print(f'TN: {TN}')
print(f'FN: {FN}')
print(f'FP: {FP}')
print(f'TP: {TP}')

# Calculate accuracy and print
accuracy = (TN + TP)/ (TN + FN + FP + TP)
print(f'Accuracy: {accuracy}')

# Calculate precision and print
precision = TP / (FP+TP)
print(f'Precision: {precision}')

# Calculate recall and print
recall = TP / (FN+TP)
print(f'Recall: {recall}')
```

```python
# Produce an ROC curve for the predicted probabilities and the Area Under the Curve (AUC) score 
# (Boorman, n.d.a)

# Predict probabilities
y_pred_proba = knn.predict_proba(X_test_scaled)[:,1]

# Calculate parameters for ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='red')
plt.plot([0, 1], [0, 1], color='blue', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()

# Calculate AUC score and print
roc_auc = auc(fpr, tpr)
print(f'AUC Score: {roc_auc}')
```
